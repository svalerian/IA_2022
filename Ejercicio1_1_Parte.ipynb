```python
def squares(numbers):
  return [number*number for number in numbers]

squares([1,2,3,4,5])
```

Esta forma de hacerlo tiene un problema, todos los resultados se generan de una vez. Si la lista numbers fuese muy grande, el tiempo de ejecucin podra ser considerable.

yield es una orden muy similar a un return, con una gran diferencia, yield pausar la ejecucin de tu funcin y guardar el estado de la misma hasta que decidas usarla de nuevo.

Y si pudiramos calcular los cuadrados uno a uno? Si no necesitamos todos los resultados, esto podra ser muy til. Ahora yield va a ser muy til. 


```python
def squares(numbers):
  for number in numbers:
    yield number*number

squares([1,2,3,4,5])
```

```python
# Ahora podemos iterar sobre el generador
for square in squares([1,2,3]):
   print(square)
```

```python
""" En este caso se genera la lista, despus se hace un return 
y por ltimo un bucle for para su impresin, con 
el consiguiente tiempo de proceso y almacenamiento"""

def genera_lista(num):
    lista = []
    i = 1
    while i <= num:
        lista.append (i)
        i += 1
    return lista

for i in genera_lista(10):
    print (i)
```

```python
""" Si utilizamos un generador nos ayudar a devolver un valor cada vez"""

def genera_lista(num):
    lista = []
    i = 1
    while i <= num:
        yield i
        i += 1
lista = genera_lista (10)

for i in genera_lista(10):
    print (i)
```

```python
def contador(max):
    print("=Dentro de contador - empezando")
    n=0
    while n < max:
        print(f"=Dentro de contador - viene yield con n={n}")
        yield n
        print("=Dentro de contador - retomando despus de yield")
        n=n+1
    print("=Dentro de contador - terminando")

print("Instanciando contador") 
mycont = contador(3)
print("Contador instanciado") 

for i in mycont:
    print(f"valor leido del iterador={i}") 
print("Listo") 
```

```python
# Primero se ejecuta la funcin saludo ("Nombre"), la salida es "Hola" + nombre
# La funcin de entrada funcion_entrada de decorador es saludo(nombre)
# El param de la funcion_salida es nombre
# Luego si el nombre es Pablo, nos devuelve la funcion_salida "Buenos das, guapo"

def decorador(funcion_entrada):
    def funcion_salida (param):
        if param == "Lidia":
            print ("Buenos das, guapa")
        elif param == "Pablo":
            print ("Buenos das, guapo")
        else:
            funcion_entrada(param)
        print ("Que tengas un buen da")
    return funcion_salida 

@decorador
def saludo(nombre):
    print ("Hola " + nombre)
    
saludo ("Pedro")
saludo ("Lidia")
```

```python
# Tenemos una funcin suma() que vamos a decorar usando mi_decorador(). 
# Para ello, antes de la declaracin de la funcin suma, hacemos uso de @mi_decorador.

def mi_decorador(funcion):
    def nueva_funcion(a, b):
        print("Se va a llamar")
        print (a, b)
        c = funcion(a, b)
        print("Se ha llamado")
        return c
    return nueva_funcion

@mi_decorador
def suma(a, b):
    print("Entra en funcion suma")
    return a + b

suma(5,8)

# Se va a llamar
# Entra en funcion suma
# Se ha llamado
```
